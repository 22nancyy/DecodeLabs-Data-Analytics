```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("dataanalytics.csv")
```

```python
print("="*60)
print("DATASET LOADED SUCCESSFULLY")
print("="*60)

print(df.head())
```

```python
conn = sqlite3.connect(":memory:")

df.to_sql(
    "orders",
    conn,
    index=False,
    if_exists="replace"
)

print("\nDatabase Created Successfully!")
```

```python
def run_query(title, query):

    print("\n")
    print("="*80)
    print(title)
    print("="*80)

    print(query)

    result = pd.read_sql_query(query, conn)

    print("-"*80)
    print(result)

    return result
```

```python


run_query(
"37. Categorize Orders Based on Total Price",

"""
SELECT
OrderID,
Product,
TotalPrice,

CASE

WHEN TotalPrice >= 2500 THEN 'Premium Order'

WHEN TotalPrice >= 1500 THEN 'High Value'

WHEN TotalPrice >= 700 THEN 'Medium Value'

ELSE 'Low Value'

END AS OrderCategory

FROM orders

ORDER BY TotalPrice DESC;

"""
)
```

```python
run_query(
"38. Count Orders By Category",

"""

SELECT

CASE

WHEN TotalPrice >=2500 THEN 'Premium Order'

WHEN TotalPrice >=1500 THEN 'High Value'

WHEN TotalPrice >=700 THEN 'Medium Value'

ELSE 'Low Value'

END AS Category,

COUNT(*) AS TotalOrders

FROM orders

GROUP BY Category;

"""
)
```
